# Testing Algoritms to use for refining the data set



```python
# First import dataset
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
import pandas as pd
df = pd.read_csv('data/raw.csv')
len(df.index)
```

```python
df.fillna(value=0.0, inplace=True)
df = pd.get_dummies(df)
```

```python
y_train = df["Operatieduur"]
y_train = y_train.values[0:3667]
df.drop(labels="Operatieduur", axis=1, inplace=True)

y_train
```

```python
X_train = df.values[0:3667]
X_test = df.values[3667:]
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

state = 12  
test_size = 0.30


X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,  
    test_size=test_size, random_state=state)
```

```python
lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=learning_rate, max_features=10, max_depth=2, random_state=0)
    gb_clf.fit(X_train, y_train)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_val, y_val)))
  
```
