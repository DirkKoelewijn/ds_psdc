```python
#######################################################################################################
# Import libraries
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, VotingRegressor, BaggingRegressor
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
from sklearn.model_selection import cross_val_score, train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
#######################################################################################################
# Best configs yet:

# model_1 = GradientBoostingRegressor(
#         loss='huber',
#         n_estimators=200,
#         max_depth=5,
#         max_features=25,
#         random_state=0,
#     )
#     model_2 = Ridge(alpha=5)
    

#     model=VotingRegressor([('GB', model_1), ('Rd', model_2)])
#     model.fit(X, y)

#     model_1 = GradientBoostingRegressor(
#         loss='huber',
#         n_estimators=200,
#         max_depth=5,
#         max_features=25,
#         random_state=0,
#     )
#     model_2 = KernelRidge(alpha=4)

#######################################################################################################


# data_set = ['../data/data-SNA1.csv', '../data/data-SNF1.csv', '../data/data-SNA3.csv', '../data/data-SNF3.csv', '../data/data-MNF1.csv', '../data/data-MNA1.csv', '../data/data-MNF3.csv', '../data/data-MNA3.csv']
data_set = ["../data/data-SNF3.csv"]


for i in data_set:
    print("DATA_SET: " + i)
    X = pd.read_csv(i)
    y = X.pop('Operatieduur').values
    
    model_GB = GradientBoostingRegressor(
        loss='huber',
        n_estimators=200,
        max_depth=5,
        max_features=25,
        random_state=0,
    )
    model_KR = KernelRidge(
        alpha=4
    )
    model_RF = RandomForestRegressor(
        n_estimators=250,
        max_features=35,
        random_state=0,
        n_jobs=-1
    )
    model_BGB = BaggingRegressor(base_estimator=model_GB,
        n_estimators=10,
        random_state=0,
        n_jobs=-1
    )
    model_BRF = BaggingRegressor(base_estimator=model_RF,
        n_estimators=10,
        random_state=0,
        n_jobs=-1
    )
    

    model=VotingRegressor([('BGB', model_BGB), ('GB', model_GB), ('KR', model_KR), ('RF', model_RF), ('BRF', model_BRF)], n_jobs=-1)
    model.fit(X, y)


    predicted = model.predict(X)
    print('\nR2:', r2_score(y, predicted))
    print('EVS:', explained_variance_score(y, predicted))
    print('MAE:', mean_absolute_error(y, predicted))
    print('MSE:', mean_squared_error(y, predicted))

    
    score = np.mean(cross_val_score(model, X, y, cv=10, scoring='r2'))
    print("\ncross validation score: ")
    print(score)
    print("\n")

    plt.figure(figsize=[8.0,8.0])
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.scatter(y, predicted)
    plt.plot(y, y, color='r')
    plt.grid(True)
    plt.show()
```
