```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score

df = pd.read_csv('data/data.csv')
df['Operatieduur'] = pd.to_numeric(df['Operatieduur'], errors='coerce')
df['Geplande operatieduur'] = pd.to_numeric(df['Geplande operatieduur'], errors='coerce')
df = df[df['Operatieduur'].notnull()]
df = df[df['Geplande operatieduur'].notnull()]
y = df.pop('Operatieduur').values
predicted = df.pop('Geplande operatieduur').values

print('\nR2:', r2_score(y, predicted))
print('EVS:', explained_variance_score(y, predicted))
print('MAE:', mean_absolute_error(y, predicted))
print('MSE:', mean_squared_error(y, predicted))

print(np.mean(np.absolute(predicted - y)))

plt.figure(figsize=[10.0,7.0])
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.scatter(y, predicted)
plt.plot(y, y, color='r')
plt.grid(True)
plt.show()

plt.figure(figsize=[10.0,7.0])
plt.xlabel('Actual')
plt.ylabel('Predicted - Actual')
plt.scatter(y, predicted - y)
plt.grid(True)
plt.show()
```
